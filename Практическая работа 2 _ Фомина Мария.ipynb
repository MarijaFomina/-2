```python
# Посмотрим на зависимость суммарного уровня дохода от уровня образования и семейного статуса

fig, axes = plt.subplots(3,2, figsize = (40, 25))
for ax, c in zip(axes.ravel(), income_by_education_status.Education.unique()):
    tmp_df = income_by_education_status[income_by_education_status.Education.eq(c)].dropna()\
    .sort_values(by = 'Income')[-10:]
    tmp_df.plot.barh(x = 'Marital_Status', ax = ax, legend = None, fontsize = 20, color = 'orangered')
    ax.set_title(c, fontsize = 30, fontweight = '600')
    plt.tight_layout()
    ax.yaxis.label.set_visible(False)
    
plt.show()
```

### Теперь посмотрим на выбросы, аномалии и построим доверительные интервалы

```python
# Таким образом, очевидно, что самые больше выбросы у переменной Income, посмотрим на них

q1 = marketing ['Income'].quantile(0.25)
q3 = marketing ['Income'].quantile(0.75)
iqr = q3 - q1

iqr_outlier_threshold_bottom = q1 - 2.5 * iqr
iqr_outlier_threshold_up = q3 + 2.5 * iqr
print(iqr_outlier_threshold_bottom)
print(iqr_outlier_threshold_up)
```

```python
marketing[marketing['Income'] > iqr_outlier_threshold_up]['Marital_Status'].value_counts()

# Также мы видим, что выбросы есть в категориях семейного статусы Together, Married, Divorced
```

```python
# Посмотрим на график доверительного интервала, выбросы четко видны

def norm_confidence(marketing = marketing, column = 'Income', alpha = 0.99):
    interval = stats.norm.interval(alpha, loc = marketing[column].mean(), scale = marketing[column].std())
    return interval
confidence = norm_confidence()
```

```python
marketing.Income.plot(figsize = (15,10))
plt.axhline(y = confidence[1], color = 'deeppink', linestyle = '-')
plt.axhline(y = confidence[0], color = 'orangered', linestyle = '-')
plt.title('Доверительный интервал Income', fontsize = 15)

plt.show()
```

```python
# Теперь посмотрим на аномальные значения данной переменной.
# Именно у нее они, скорее всего, наиболее критичные 

def find_anomaly(marketing = marketing, column = 'Income'): 
    upper_level = marketing[column].mean() + 3 * marketing[column].std()
    lower_level = marketing[column].mean() - 3 * marketing[column].std()
    anomaly = marketing[(marketing[column] > upper_level) & (marketing[column] < lower_level)]
    return upper_level, lower_level, anomaly
```

```python
anomaly = find_anomaly(marketing = marketing, column = 'Income')
anomaly
```

> Таким образом, мы выявили выбросы и аномалии данной переменной. Их стоит удалить, так как они могут помешать дальнейшему анализу корреляции и регресии. 

```python
dev = (marketing['Income'] - marketing['Income'].mean()).abs()

std = marketing['Income'].std()

# Фильтруем исходный набор данных
marketing_filtering = marketing[ ~ (dev > 3 * std)]

plt.figure(figsize = (20,10))
plt.title('Распределение доходов без выбросов', fontsize = 15) 
plt.tick_params(labelsize = 15, color = 'orangered')
plt.xticks(np.arange(0, max(marketing['Income']), 7000))


sns.distplot(marketing_filtering['Income'], axlabel = False, bins = 50)

plt.show()
```

### Теперь поставим гипотезы, которые мы будем проверять:

### H0: Отсутсвует взаимосвять между уровнем образования и уровнем дохода 
### H1: Существует статистически значимая взаимосвязь между уровнем образования и уровнем дохода

```python
ct_income = marketing_filtering.groupby('Income_cat1')['Education'].value_counts().unstack()
ct_income
```

```python
# Избавимся от значений типа NaN, так как они помешают посчитать нам критерии статистики

ct_income_without_NaN = ct_income.fillna(ct_income.mean())
```

```python
# Таким образом, мы видим значения хи-квадрата, степеней свободы и p-value 

chi_2, p_value, df, expected = scipy.stats.chi2_contingency(ct_income_without_NaN, correction = False)
output = "chi_2: {}\ndegrees of freedom: {}\np-value: {}\n"
print(output.format(chi_2,df, p_value))
```

```python
# Зададим значение альфа, а также кол-ва степеней свободы для подсчета критического значения

alpha = 0.05 
critical_value = scipy.stats.chi2.ppf(0.95, 20)
critical_value
```

```python
# В итоге, проверяем наши гипотезы двумя способами: 
# Если хи-квадрат больше или равен критическому значению, то мы отвергаем нулевую гипотезу 
# Если p-value меньше альфа, то мы аналогично отвергаем гипотезу

if chi_2 >= critical_value:
    print('Отвергаем Н0')
else:
    print('Не отвергаем Н0')
```

```python
if p_value < alpha:
    print('Отвергаем Н0')
else:
    print('Не отвергаем Н0')
```

### Таким образом, мы проверили двумя способами нашу гипотезу. В итоге мы отвергаем H0, которая гласит, что "Отсутсвует взаимосвзять между уровнем образования и уровнем дохода", следовательно статистическая значимость между взаимосвязью уровнем образования и уровнем дохода существует.
